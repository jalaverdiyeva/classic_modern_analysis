```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax

tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-xlm-roberta-base-sentiment")
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-xlm-roberta-base-sentiment")

def get_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model(**inputs)
    scores = softmax(outputs.logits.detach().numpy()[0])
    labels = ["Negative", "Neutral", "Positive"]
    label = labels[scores.argmax()]
    return scores.max(), label

df["Sentiment_Score"], df["Sentiment_Label"] = zip(*df["Modern_Sentence"].apply(get_sentiment))
```

```python
from sentence_transformers import SentenceTransformer, util

# Load model
model_sim = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
embeddings = model_sim.encode(df["Modern_Sentence"].tolist(), convert_to_tensor=True)

# Example: find most similar sentence for the first sentence
query_emb = embeddings[0]
scores = util.cos_sim(query_emb, embeddings)[0]
most_sim_idx = scores.argmax().item()  # convert tensor to integer
similarity_score = scores.max()

df.loc[0, "Similarity_With_Query"] = similarity_score.item()
df.loc[0, "Most_Similar_Sentence_ID"] = df.loc[most_sim_idx, "ID"]
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
from transformers import pipeline

# Load your Excel file
df = pd.read_excel("classic_modern_analysis.xlsx")

# ------------------------
# 1️⃣ Sentiment Analysis
# ------------------------
# Using multilingual BERT sentiment pipeline
sentiment_analyzer = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Fill Sentiment_Score and Sentiment_Label for Classic_Sentence
classic_sentiments = sentiment_analyzer(df["Classic_Sentence"].tolist())
df["Sentiment_Score"] = [s["score"] for s in classic_sentiments]
df["Sentiment_Label"] = [s["label"] for s in classic_sentiments]

# Fill mbert_mod_label for Modern_Sentence
modern_sentiments = sentiment_analyzer(df["Modern_Sentence"].tolist())
df["mbert_mod_label"] = [s["label"] for s in modern_sentiments]

# ------------------------
# 2️⃣ Sentence Similarity
# ------------------------
# Load Sentence-BERT model
model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# Encode all Classic sentences
classic_embeddings = model.encode(df["Classic_Sentence"].tolist(), convert_to_tensor=True)
modern_embeddings = model.encode(df["Modern_Sentence"].tolist(), convert_to_tensor=True)

# Calculate similarity and most similar sentence
similarities = []
most_similar_ids = []

for i, emb in enumerate(classic_embeddings):
    # Compute cosine similarity with all modern embeddings
    sim_scores = util.cos_sim(emb, modern_embeddings)[0]  # tensor of similarities
    max_idx = sim_scores.argmax().item()  # index of most similar modern sentence
    similarities.append(sim_scores[max_idx].item())
    most_similar_ids.append(df.loc[max_idx, "ID"])

df["Similarity_With_Query"] = similarities
df["Most_Similar_Sentence_ID"] = most_similar_ids

# ------------------------
# 3️⃣ Notes column
# ------------------------
notes = []
for i, row in df.iterrows():
    note = ""
    # Check if sentiment differs
    if row["Sentiment_Label"] != row["mbert_mod_label"]:
        note += "Sentiment differs. "
    # Check if similarity is low
    if row["Similarity_With_Query"] < 0.6:
        note += "Low semantic similarity. "
    # Optional: add manual comment if modern sentence simplifies meaning
    if len(row["Classic_Sentence"]) - len(row["Modern_Sentence"]) > 20:
        note += "Modern sentence simplified. "
    notes.append(note.strip() if note else "OK")

df["Notes"] = notes

# ------------------------
# 4️⃣ Save updated Excel
# ------------------------
df.to_excel("classic_modern_analysis_filled.xlsx", index=False)
print("✅ All columns filled and saved to classic_modern_analysis_filled.xlsx")
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Assuming labels are like '1 star', '2 stars', ... '5 stars'
y_true = df["Sentiment_Label"]
y_pred = df["mbert_mod_label"]

precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
print(f"F1-score: {f1:.3f}")
```
